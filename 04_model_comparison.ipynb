```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor


plt.style.use("ggplot")


processed_df = pd.read_csv("raw_data/cleaned_data.csv")

split_idx = int(len(processed_df) * 0.96)

X = processed_df.drop(columns=["ClosePrice"])
y = processed_df["ClosePrice"]

X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

baseline_test_r2 = 0.7926
```

```python
# Decision Tree 
dt_model = DecisionTreeRegressor(max_depth = 15, random_state = 42)
dt_model.fit(X_train, y_train)

# predictions
dt_train_pred = dt_model.predict(X_train)
dt_test_pred = dt_model.predict(X_test)

# evaluate 
dt_train_r2 = r2_score(y_train, dt_train_pred)
dt_test_r2 = r2_score(y_test, dt_test_pred)

print("--- Decision Tree Results ---")
print(f"Training R² Score: {dt_train_r2:.4f}")
print(f"Testing R² Score:  {dt_test_r2:.4f}")
print(f"vs. Baseline OLS:  {dt_test_r2 - baseline_test_r2:+.4f}")
```

```python
# Random Forest; n_jobs = -1 to speed it up  
rf_model = RandomForestRegressor(n_estimators=50, max_depth=12, n_jobs=-1, random_state=42)
rf_model.fit(X_train, y_train)

# predict 
rf_train_pred = rf_model.predict(X_train)
rf_test_pred = rf_model.predict(X_test)

# evaluate 
rf_train_r2 = r2_score(y_train, rf_train_pred)
rf_test_r2 = r2_score(y_test, rf_test_pred)

print("--- Random Forest Results ---")
print(f"Training R² Score: {rf_train_r2:.4f}")
print(f"Testing R² Score:  {rf_test_r2:.4f}")
print(f"vs. Baseline OLS:  {rf_test_r2 - baseline_test_r2:+.4f}")
```

The regression model outperformed both tree models. One reason is because of the OHE columns. OLS assigns continuous weights but decision trees will split the features, thus leaving the data very sparse since only a small fraction of data will belong to a certain city. The tree model also can't extrapolate outside the training data.
